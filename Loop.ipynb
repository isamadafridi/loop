# **What is a Loop in Programming?**

A loop is a way to repeat a block of code multiple times

Helps automate repetitive tasks instead of writing code over and over

Two main types in Python:

  -**for loop** – iterates over sequences (like lists, strings, etc.)

  -**while loop** – repeats as long as a condition is True

Loops are used in data processing, automation, and NLP tasks (e.g., iterating over words)

# **Why Loops are important?**

**Reduce code length and complexity:**
Loops let you do more with less code by avoiding repetition.

**Make programs efficient and dynamic:**
They adapt to data size and user input, running actions as needed.

**Essential for handling collections of data:**
Loops process each item in lists, sentences, or datasets automatically.

# **For Loop**

-A for loop is used to iterate over a sequence (like
a list, string, tuple, or dictionary).

-It automatically goes through each item in the
sequence one by one.

**Syntax:**

    for item in sequence:

    # do something with item

**Example 1: Looping through a list**

```python
words = ["hello", "python", "nlp"]

for word in words:

  print(word.upper())
```

**Example 2: Looping through a string**

```python
for char in "text":

  print(char)
```

**Common Use Cases in NLP**

```python
sentence = "Python makes Natural Language Processingfun and powerful"

words = sentence.lower().split()
 #output; words : ['python', 'makes', 'natural', 'language', 'processingfun', 'and', 'powerful']

count = 0

for word in words:

  if len(word) > 4:

    count += 1

print("Number of words longer than 4 characters:", count)
```

```python
sentence = "NLP makes NLP fun and powerful"

words = sentence.lower().split()

freq = {} # Empty dictionary to store frequencies

for word in words:

  freq[word] = freq.get(word, 0) + 1

print(freq)
```

# **While Loop**

A while loop repeats a block of code as long as a
given condition is True

Great for repeating until a condition changes

**Syntax:**

    while condition:

      # do something

Be Careful with Infinite Loops

If the condition never becomes False, the loop will run
forever

**When to Use while in NLP**

Waiting for specific input in chatbots

Looping until text matches a pattern

Reading lines from a file until EOF

Interactive tools or pipelines with manual breakpoints

# **Break**

break – Exit the Loop Early

Immediately stops the loop, even if the condition is still true

Used when a certain condition is met or an error occurs

Example:



```python
for word in ["hello", "nlp", "stop", "python"]:

      if word == "stop":

          break

      print(word)
```

# **continue**

– Skip Current Iteration

Skips the current loop cycle, and moves to the next

Useful for ignoring unwanted items (e.g., stopwords in NLP)

Example:

   

```python
for word in ["the", "is", "awesome", "in"]:

    if word in ["the", "is", "in"]:

        continue

    print(word)
```

# **pass**

– Do Nothing (Placeholder)

Does nothing at all — just a placeholder to avoid syntax errors

Useful during development or to define a loop or function structure

Example:

```python
for word in ["token", "skip", "here"]:

    if word == "skip":

        pass # we'll handle it later

    print(word)
```

# **Nested Loops**

A loop inside another loop

Inner loop completes all iterations in each iteration of the outer loop

Allows us to iterate through multi-level data structures

Common when working with lists of lists, matrices, or token-tag pairs in NLP

Example:

    for i in range(3):

        for j in range(2):

          print(f"i={i}, j={j}")


**While using nested loops:**

Keep nesting levels low (max 2–3)

Use meaningful variable names (word, sentence, not i, j)

Avoid unnecessary nesting → can hurt readability & performance

Break into functions when complexity increases

# **List Comprehensions in Python**

A concise way to create lists using a single line of code

Replaces for loops for generating new lists

Syntax:

[expression for item in iterable if condition]

Example:

```python
numbers = [1, 2, 3, 4, 5]

squares = [x ** 2 for x in numbers]

print(squares) # [1, 4, 9, 16, 25]
```

```python
# Example:

sentence = "Python is powerful and simple"

words = sentence.split()

short_words = [w for w in words if len(w) <= 4]

print(short_words) # ['is', 'and']
```

```python
# Example:

sentence = "Python is powerful and simple"

words = sentence.split()

word_len_pairs = [(w, len(w)) for w in words]

print(word_len_pairs)
```

```python
# Example: Nested loops

matrix = [[1, 2], [3, 4], [5, 6]]

flattened = [item for row in matrix for item in row]

print(flattened) # [1, 2, 3, 4, 5, 6]
```

# **While using nested loops in list comprehensions:**

Use for simple, readable logic

Avoid if logic is too complex → stick to for loops

Works with any iterable (strings, lists, tuples, etc.)

## **What is an iterable?**

An object you can loop over (like in a for loop)

Examples: list, tuple, set, dict, str

Implements the __iter__() method

   

```python
words = ["nlp", "python", "fun"] # ← Iterable

for word in words:

    print(word)
```

# **What is an iterator?**

An object that remembers its state while iterating

Has two key methods:

__iter__() → returns the iterator itself

__next__() → returns the next item or raises
StopIteration

    words = ["token", "model", "vector"]

    it = iter(words)
Convert to iterator

    print(next(it))
'token'

    print(next(it))
'model'

# **✅ Example: Custom Iterator for a Countdown**

```python
class Countdown:
    def __init__(self, start):
        self.current = start

    def __iter__(self):
        # __iter__ must return the iterator object itself
        return self

    def __next__(self):
        # __next__ defines how to get the next item
        if self.current <= 0:
            raise StopIteration
        else:
            num = self.current
            self.current -= 1
            return num

# Using the iterator
cd = Countdown(5)

for number in cd:
    print(number)
```
